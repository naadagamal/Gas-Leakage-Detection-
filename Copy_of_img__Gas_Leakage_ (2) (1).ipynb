# <span style='color:chocolate'>Data Profiling</span>

```python
def data_profiling(Gas_Sensors):
    data_profile = []
    columns = Gas_Sensors.columns
    for col in columns:
        unique = Gas_Sensors[col].nunique()
        NaN = Gas_Sensors[col].isnull().sum()
        duplicates = Gas_Sensors[col].duplicated().sum()
        data_profile.append([col,unique,NaN,duplicates])
    table = pd.DataFrame(data_profile)
    table.columns = ['column','unique','NaN','duplicates']
    return table
```

```python
import cv2
import numpy as np

# Load the Haar-like feature classifier
face_cascade = cv2.CascadeClassifier('haarcascade_frontalface_default.xml')

# Open the video stream
try:
    cap = cv2.VideoCapture(0)
except cv2.error as e:
    print(f"Error: {e}")
    exit()

fourcc = cv2.VideoWriter_fourcc(*'XVID')  # Choose the codec
out = cv2.VideoWriter('c:/Videos/record/out.avi', fourcc, 20.0, (640, 480))  # Create a VideoWriter object

counter = 0

while True:
    # Read a frame from the video stream
    ret, img = cap.read()

    # Check if the frame was read successfully
    if not ret:
        print("Error: Unable to read frame from video stream")
        break

    # Convert the frame to grayscale
    gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

    # Detect faces in the frame
    faces = face_cascade.detectMultiScale(gray, 1.1, 4)

    # Draw rectangles around the detected faces and count the number of people
    for (x, y, w, h) in faces:
        cv2.rectangle(img, (x, y), (x + w, y + h), (255, 0, 0), 3)
        counter = faces.shape[0]

    # Display the number of people detected
    cv2.putText(img, f"Number of people: {counter}", (10, 30), cv2.FONT_HERSHEY_SIMPLEX, 1, (0, 0, 255), 2)

    # Write the frame to the output file
    out.write(img)

    # Display the output
    cv2.imshow('Video Stream', img)

    # Exit the loop if the user presses the 'q' key
    if cv2.waitKey(1) == ord('q'):
        break

# Print the number of people detected
print(f"Number of people detected: {counter}")

# Release the video stream, close the windows, and release the VideoWriter object
cap.release()
cv2.destroyAllWindows()
out.release()
```
